## Preprocessing

```python
from sklearn.compose import LabelEncoderEncoder
from sklearn.preprocessing import MinMaxScaler
```

```python
X=df.drop("Loan_status",axis=1)
y=df["Loan_status"]

le=LabelEncoder()
df['Self_employed']=le.fit_transform(df['Self_employed'])
df['Eduaciton']=le.fit_transform(df['Eduaciton'])
df['Loan_status']=le.fit_transform(df['Loan_status'])

```

```python
minMaxScaler = MinMaxScaler().fit(X)  
X_processed = minMaxScaler.transform(X)
```

```python
df_processed = pd.DataFrame(X_processed, columns=X.columns)
df_processed
```

```python
sns.boxplot(data=df_processed)
plt.xticks(rotation=45)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_processed, y, test_size=0.2, random_state=42)

print(X_train.shape)
X_test.shape
```

## Training & evaluation

#### We will use classification model to train and evaluation

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
#Find out which one model is better
```

```python
import collections, operator
Runresult=collections.namedtuple("Runresult",'model pred accuracy clf_report conf_mat')

                                 
rand_state = 42
model_list=[
            SVC(kernel='linear'),
            SVC(kernel='rbf'),
            SVC(kernel='poly'),
            SVC(kernel='sigmoid'),
            KNeighborsClassifier(n_neighbors=5),
            LogisticRegression(max_iter=5000),
            DecisionTreeClassifier(),
            RandomForestClassifier(),
           ]

result_list=[]

for i in model_list:
    i.fit(X_train,y_train)
    pred=i.predict(X_test)
    accuracy=accuracy_score(y_test,pred)
    clf_report=classification_report(y_test,pred)
    conf_mat=confusion_matrix(y_test,pred)

    result_list.append(Runresult(i, pred, accuracy, clf_report, conf_mat))

```

```python
result_list.sort(key=operator.attrgetter('accuracy'),reverse=True)

space=" "
sns.set(font_scale=0.9)
fig,axes=plt.subplots(2,4,figsize=(15,8))
axes=axes.ravel()

for i ,result in enumerate(result_list):
    model_name=type(result.model).__name__
    print(f"{i+1}. {model_name:30}                   {result.accuracy:6.5f}")
    print("-"*55)
    print(f'  {result.clf_report}')

    # sns.heatmap(result.conf_mat,square=True,annot=True,fmt='d',camp='Reds',ax=axes[i])
    # axes[i].set_title(model_name)
    sns.heatmap(result.conf_mat, square=True, annot=True, fmt='d', cmap='Reds', ax=axes[i])
    axes[i].set_title(model_name)
    axes[i].set_xlabel("Predict")
    axes[i].set_ylabel("Actual")

best_model=result_list[0].model
```
